```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
Fever=ctrl.Antecedent(np.arange(0,11,1),"Fever")
Headache=ctrl.Antecedent(np.arange(0,11,1),"Headache")
Nausea=ctrl.Antecedent(np.arange(0,11,1),"Nausea")
Vomiting=ctrl.Antecedent(np.arange(0,11,1),"Vomiting")
Jaundice=ctrl.Antecedent(np.arange(0,11,1),"Jaundice")
Jointpain=ctrl.Antecedent(np.arange(0,11,1),"Jointpain")
Bodyweekness=ctrl.Antecedent(np.arange(0,11,1),"Bodyweekness")

Typhoid=ctrl.Consequent(np.arange(0,11,1),"Typhoid")

Fever["Mild"]=fuzzy.trimf(Fever.universe,[0,2,4])
Fever["Moderate"]=fuzzy.trimf(Fever.universe,[3,6,8])
Fever["Severe"]=fuzzy.trimf(Fever.universe,[2,3,5])
Fever["Very Severe"]=fuzzy.trimf(Fever.universe,[6,8,10])
Fever["None"]=fuzzy.trimf(Fever.universe,[7,10,10])

Headache["Mild"]=fuzzy.trimf(Headache.universe,[0,2,4])
Headache["Moderate"]=fuzzy.trimf(Headache.universe,[3,6,8])
Headache["Severe"]=fuzzy.trimf(Headache.universe,[2,3,5])
Headache["Very Severe"]=fuzzy.trimf(Headache.universe,[6,8,10])
Headache["None"]=fuzzy.trimf(Headache.universe,[7,10,10])

Nausea["Mild"]=fuzzy.trimf(Nausea.universe,[0,2,4])
Nausea["Moderate"]=fuzzy.trimf(Nausea.universe,[3,6,8])
Nausea["Severe"]=fuzzy.trimf(Nausea.universe,[2,3,5])
Nausea["Very Severe"]=fuzzy.trimf(Nausea.universe,[6,8,10])
Nausea["None"]=fuzzy.trimf(Nausea.universe,[7,10,10])

Vomiting["Mild"]=fuzzy.trimf(Vomiting.universe,[0,2,4])
Vomiting["Moderate"]=fuzzy.trimf(Vomiting.universe,[3,6,8])
Vomiting["Severe"]=fuzzy.trimf(Vomiting.universe,[2,3,5])
Vomiting["Very Severe"]=fuzzy.trimf(Vomiting.universe,[6,8,10])
Vomiting["None"]=fuzzy.trimf(Vomiting.universe,[7,10,10])

Jaundice["Mild"]=fuzzy.trimf(Jaundice.universe,[0,2,4])
Jaundice["Moderate"]=fuzzy.trimf(Jaundice.universe,[3,6,8])
Jaundice["Severe"]=fuzzy.trimf(Jaundice.universe,[2,3,5])
Jaundice["Very Severe"]=fuzzy.trimf(Jaundice.universe,[6,8,10])
Jaundice["None"]=fuzzy.trimf(Jaundice.universe,[7,10,10])

Jointpain["Mild"]=fuzzy.trimf(Jointpain.universe,[0,2,4])
Jointpain["Moderate"]=fuzzy.trimf(Jointpain.universe,[3,6,8])
Jointpain["Severe"]=fuzzy.trimf(Jointpain.universe,[2,3,5])
Jointpain["Very Severe"]=fuzzy.trimf(Jointpain.universe,[6,8,10])
Jointpain["None"]=fuzzy.trimf(Jointpain.universe,[7,10,10])

Bodyweekness["Mild"]=fuzzy.trimf(Bodyweekness.universe,[0,2,4])
Bodyweekness["Moderate"]=fuzzy.trimf(Bodyweekness.universe,[3,6,8])
Bodyweekness["Severe"]=fuzzy.trimf(Bodyweekness.universe,[2,3,5])
Bodyweekness["Very Severe"]=fuzzy.trimf(Bodyweekness.universe,[6,8,10])
Bodyweekness["None"]=fuzzy.trimf(Bodyweekness.universe,[7,10,10])

Typhoid["Mild"]=fuzzy.trimf(Typhoid.universe,[0,2,4])
Typhoid["Moderate"]=fuzzy.trimf(Typhoid.universe,[3,6,8])
Typhoid["Severe"]=fuzzy.trimf(Typhoid.universe,[2,3,5])
Typhoid["Very Severe"]=fuzzy.trimf(Typhoid.universe,[6,8,10])
Typhoid["None"]=fuzzy.trimf(Typhoid.universe,[7,10,10])

Fever.view()
Headache.view()
Nausea.view()
Vomiting.view()
Jaundice.view()
Jointpain.view()
Bodyweekness.view()

Typhoid.view()
plt.show()

Rule1=ctrl.Rule(Fever["Mild"] & Headache["Mild"] & Nausea["Mild"] & Vomiting["Mild"] & Jaundice["Mild"]
                & Jointpain["Mild"] & Bodyweekness["Mild"], Typhoid["Mild"])

Rule2=ctrl.Rule(Fever["Moderate"] & Headache["Moderate"] & Nausea["Moderate"] & Vomiting["Moderate"] & Jaundice["Moderate"]
                & Jointpain["Moderate"] & Bodyweekness["Moderate"], Typhoid["Moderate"])

Rule3=ctrl.Rule(Fever["Severe"] & Headache["Severe"] & Nausea["Severe"] & Vomiting["Severe"] & Jaundice["Severe"]
                & Jointpain["Severe"] & Bodyweekness["Severe"], Typhoid["Severe"])

Rule4=ctrl.Rule(Fever["Very Severe"] & Headache["Very Severe"] & Nausea["Very Severe"] & Vomiting["Very Severe"] & Jaundice["Very Severe"]
                & Jointpain["Very Severe"] & Bodyweekness["Very Severe"], Typhoid["Very Severe"])

Rule5=ctrl.Rule(Fever["None"] & Headache["None"] & Nausea["None"] & Vomiting["None"] & Jaundice["None"]
                & Jointpain["None"] & Bodyweekness["None"], Typhoid["None"])

Rule6=ctrl.Rule(Fever["Moderate"] & Headache["Mild"] & Nausea["Mild"] & Vomiting["Mild"] & Jaundice["Severe"]
                & Jointpain["Very Severe"] & Bodyweekness["Severe"], Typhoid["Moderate"])

Rule7=ctrl.Rule(Fever["Mild"] & Headache["Moderate"] & Nausea["Mild"] & Vomiting["Moderate"] & Jaundice["Mild"]
                & Jointpain["Moderate"] & Bodyweekness["Moderate"], Typhoid["Moderate"])

Rule8=ctrl.Rule(Fever["Mild"] & Headache["Mild"] & Nausea["Moderate"] & Vomiting["Mild"] & Jaundice["Severe"]
                & Jointpain["Mild"] & Bodyweekness["Mild"], Typhoid["Mild"])

Rule9=ctrl.Rule(Fever["Severe"] & Headache["Severe"] & Nausea["Mild"] & Vomiting["Moderate"] & Jaundice["Mild"]
                & Jointpain["Very Severe"] & Bodyweekness["Moderate"], Typhoid["Severe"])

Rule10=ctrl.Rule(Fever["Moderate"] & Headache["Severe"] & Nausea["Very Severe"] & Vomiting["Very Severe"] & Jaundice["Mild"]
                & Jointpain["Severe"] & Bodyweekness["Very Severe"], Typhoid["Severe"])

Rule11=ctrl.Rule(Fever["Very Severe"] & Headache["Very Severe"] & Nausea["Severe"] & Vomiting["Very Severe"] & Jaundice["Mild"]
                & Jointpain["Severe"] & Bodyweekness["Very Severe"], Typhoid["Very Severe"])

Rule12=ctrl.Rule(Fever["Moderate"] & Headache["Mild"] & Nausea["None"] & Vomiting["Mild"] & Jaundice["Moderate"]
                & Jointpain["Mild"] & Bodyweekness["Mild"], Typhoid["Mild"])

Rule13=ctrl.Rule(Fever["Severe"] & Headache["Moderate"] & Nausea["Mild"] & Vomiting["Very Severe"] & Jaundice["Moderate"]
                & Jointpain["Very Severe"] & Bodyweekness["Very Severe"], Typhoid["Very Severe"])

Rule14=ctrl.Rule(Fever["Very Severe"] & Headache["Moderate"] & Nausea["Mild"] & Vomiting["Moderate"] & Jaundice["Moderate"]
                & Jointpain["Severe"] & Bodyweekness["Moderate"], Typhoid["Moderate"])

Rule15=ctrl.Rule(Fever["Moderate"] & Headache["None"] & Nausea["Mild"] & Vomiting["None"] & Jaundice["Mild"]
                & Jointpain["None"] & Bodyweekness["None"], Typhoid["None"])


Fev_ctrl = ctrl.ControlSystem([Rule1,Rule2,Rule3,Rule4,Rule5,Rule6,Rule7,Rule8,Rule9,Rule10,Rule11,Rule12,Rule13,Rule14,Rule15])
Fev=ctrl.ControlSystemSimulation(Fev_ctrl)

Fev.input["Fever"]=10
Fev.input["Headache"]=10
Fev.input["Nausea"]=10
Fev.input["Vomiting"]=10
Fev.input["Jaundice"]=10
Fev.input["Jointpain"]=10
Fev.input["Bodyweekness"]=10

Fev.compute()
print(Fev.output["Typhoid"])
Typhoid.view(sim=Fev)
```
